```python
#Name: Sisham Maharjan
#Student ID: 2409266

import numpy as np
import matplotlib.pyplot as plt

def plot_function(func, a, b):
    x = np.linspace(a, b, 400)
    y = func(x)
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Graph of the function')
    plt.grid(True)
    plt.show()

def midpoint_approx(func, a, b, N):
    h = (b - a) / N
    result = 0
    for i in range(N):
        xi = a + (i + 0.5) * h
        result += func(xi)
    result *= h
    return result

if __name__ == "__main__":
    # Define the 1st function to be integrated
    func_1 = lambda x: x / (x**2 + 1)

    # Define the 2nd function to be integrated
    func_2 = lambda x: np.exp(x)

    # Plot the graph of the 1st function
    plot_function(func_1, 0, 5)

    # End points and number of partitions for the 1st function
    a_1 = 0
    b_1 = 5
    N_1 = 50

    # Print the subinterval width for the 1st function
    print("Subinterval width for 1st Function = {:.6f}".format((b_1 - a_1) / N_1))

    # Compute the midpoint approximation for the 1st function
    midpoint_approx_1 = midpoint_approx(func_1, a_1, b_1, N_1)

    # Calculate the true value of the definite integral for the 1st function
    antiderivative_1 = lambda x: 0.5 * np.log(x**2 + 1)
    definite_integral_1 = antiderivative_1(b_1) - antiderivative_1(a_1)

    # Calculating the absolute error for the 1st function
    error_1 = np.abs(midpoint_approx_1 - definite_integral_1)

    # Printing the results for the 1st function
    print("\nMidpoint Approximation for 1st Function = {:.6f}".format(midpoint_approx_1))
    print("Actual Value for 1st Function = {:.6f}".format(definite_integral_1))
    print("Absolute error for 1st Function = {:.8f}".format(error_1))

    # Plotting the graph of the 2nd function
    plot_function(func_2, 0, 5)

    # Ending points and number of partitions for the 2nd function
    a_2 = 0
    b_2 = 5
    N_2 = 50

    # Printing the subinterval width for the 2nd function
    print("Subinterval width for 2nd Function = {:.6f}".format((b_2 - a_2) / N_2))

    # Computing the midpoint approximation for the 2nd function
    midpoint_approx_2 = midpoint_approx(func_2, a_2, b_2, N_2)

    # Calculating the true value of the definite integral for the 2nd function
    antiderivative_2 = lambda x: np.exp(x)
    definite_integral_2 = antiderivative_2(b_2) - antiderivative_2(a_2)

    # Calculating the absolute error for the 2nd function
    error_2 = np.abs(midpoint_approx_2 - definite_integral_2)

    # Printing the results for the 2nd function
    print("\nMidpoint Approximation for 2nd Function = {:.6f}".format(midpoint_approx_2))
    print("Actual Value for 2nd Function = {:.6f}".format(definite_integral_2))
    print("Absolute error for 2nd Function = {:.8f}".format(error_2))
```
