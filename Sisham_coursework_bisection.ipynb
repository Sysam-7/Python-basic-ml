```python
#Name: Sisham Maharjan
#Student ID: 2409266

# Using this function f(x)=x**2 - x - 1
import math
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

def plot_function(func, a, b):
    """
    This function plots the graph of the input func
    within the given interval [a,b).
    """
    x = np.linspace(a, b, 400)
    y = func(x)
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Graph of the function')
    plt.grid(True)
    plt.show()

def bisection_method(func, a, b, tol=1e-6, max_iter=100):
    """
    Bisection method to find the root of a function within a given interval.

    Parameters:
    - func: The function for which the root is to be found.
    - a, b: Interval [a, b] within which the root is searched for.
    - tol: Tolerance level for checking convergence of the method.
    - max_iter: Maximum number of iterations.

    Returns:
    - root: Approximation of the root.
    """
    # Check if the interval is valid (signs of f(a) and f(b) are different)
    if np.sign(func(a)) * np.sign(func(b)) >= 0:
        raise ValueError("The signs of f(a) and f(b) must be different.")

    # Main loop starts here
    iter_count = 1
    while iter_count <= max_iter:
        c = (a + b) / 2  # new midpoint
        if np.abs(func(c)) < tol or (b - a) <= tol:  # solution found
            return c
        iter_count += 1
        if np.sign(func(c)) == np.sign(func(a)):
            a = c
        else:
            b = c

    print("Warning! Exceeded the maximum number of iterations.")
    return c

# Example usage:
if __name__ == "__main__":
    # Define the function for which the root is to be found
    func = lambda x: x**2 - x - 1  # First Function

    # Call plot_function to plot graph of the function
    plot_function(func, -1, 2)

    # Set the interval [a, b] for the search
    a_1 = 1; b_1 = 2  # For first root (change the values as required)
    a_2 = -1; b_2 = 0  # For second root (change the values as required)

    # Call the bisection method
    our_root_1 = bisection_method(func, a_1, b_1)
    our_root_2 = bisection_method(func, a_2, b_2)

    # Call SciPy method root, which we consider as a reference method.
    sp_result_1 = sp.optimize.root(func, (a_1 + b_1) / 2)
    sp_root_1 = sp_result_1.x.item()

    sp_result_2 = sp.optimize.root(func, (a_2 + b_2) / 2)
    sp_root_2 = sp_result_2.x.item()

    # Print the result
    print("1st root found by Bisection Method = {:0.8f}.".format(our_root_1))
    print("1st root found by SciPy = {:0.8f}".format(sp_root_1))

    print("2nd root found by Bisection Method = {:0.8f}.".format(our_root_2))
    print("2nd root found by SciPy = {:0.8f}".format(sp_root_2))




#Using this function f(x)=x**3 - x**2 - 2*x + 1
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt


def plot_function(func, a, b):
    """
    This function plots the graph of the input func
    within the given interval [a, b).
    """
    x = np.linspace(a, b, 400)
    y = func(x)
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Graph of the function')
    plt.grid(True)
    plt.show()


def bisection_method(func, a, b, tol=1e-6, max_iter=100):
    """
    Bisection method to find the root of a function within a given interval.

    Parameters:
    - func: The function for which the root is to be found.
    - a, b: Interval [a, b] within which the root is searched for.
    - tol: Tolerance level for checking convergence of the method.
    - max_iter: Maximum number of iterations.

    Returns:
    - root: Approximation of the root.
    """
    # Check if the interval is valid (signs of f(a) and f(b) are different)
    if np.sign(func(a)) * np.sign(func(b)) >= 0:
        raise ValueError("The signs of f(a) and f(b) must be different.")

    # Main loop starts here
    iter_count = 1
    while iter_count <= max_iter:
        c = (a + b) / 2  # new midpoint
        if np.abs(func(c)) < tol or (b - a) <= tol:  # solution found
            return c
        iter_count += 1
        if np.sign(func(c)) == np.sign(func(a)):
            a = c
        else:
            b = c

    print("Warning! Exceeded the maximum number of iterations.")
    return c


# Define the function for which the root is to be found
func = lambda x: x**3 - x**2 - 2*x + 1

# Call plot_function to plot graph of the function
plot_function(func, -2, 2)

# Set the intervals [a, b] for the search of each root
a_1, b_1 = -2, -1  # For the first root
a_2, b_2 = -1.5, -0.5   # For the second root
a_3, b_3 = 0.5, 2    # For the third root

# Find the roots using the bisection method
our_root_1 = bisection_method(func, a_1, b_1)
our_root_2 = bisection_method(func, a_2, b_2)
our_root_3 = bisection_method(func, a_3, b_3)

# Call SciPy method root, which we consider as a reference method.
sp_root_1 = sp.optimize.root(func, (a_1 + b_1) / 2).x.item()
sp_root_2 = sp.optimize.root(func, (a_2 + b_2) / 2).x.item()
sp_root_3 = sp.optimize.root(func, (a_3 + b_3) / 2).x.item()

# Print the results
print("1st root found by Bisection Method = {:.8f}".format(our_root_1))
print("1st root found by SciPy = {:.8f}".format(sp_root_1))

print("2nd root found by Bisection Method = {:.8f}".format(our_root_2))
print("2nd root found by SciPy = {:.8f}".format(sp_root_2))

print("3rd root found by Bisection Method = {:.8f}".format(our_root_3))
print("3rd root found by SciPy = {:.8f}".format(sp_root_3))

```
