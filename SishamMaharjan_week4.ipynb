```python

import matplotlib.pyplot as plt

y_17 = [8, 150, 80]
y_18 = [54, 77, 54]
y_19 = [93, 32, 100]
y_20 = [116, 11, 76]
y_21 = [137, 6, 93]
y_22 = [184, 1, 72]
groups = ["Bears", "Dolphins", "Whales"]

plt.subplot(2,3,1)
plt.pie(y_17, labels=groups, textprops={'fontsize' : 10})
plt.title("Population of 2017", fontsize = 10)

plt.subplot(2,3,2)
plt.pie(y_18, labels=groups, textprops={'fontsize' : 10})
plt.title("Population of 2018", fontsize = 10)

plt.subplot(2,3,3)
plt.pie(y_19, labels=groups, textprops={'fontsize' : 10})
plt.title("Population of 2019", fontsize = 10)

plt.subplot(2,3,4)
plt.pie(y_20, labels=groups, textprops={'fontsize' : 10})
plt.title("Population of 2020", fontsize = 10)

plt.subplot(2,3,5)
plt.pie(y_21, labels=groups, textprops={'fontsize' : 10})
plt.title("population of 2021", fontsize = 10)

plt.subplot(2,3,6)
plt.pie(y_22, labels=groups, textprops={'fontsize' : 10})
plt.title("Population of 2022", fontsize = 10)

plt.tight_layout(pad=1)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Red", "Silver", "Blue", "White", "Green"]
y = [123, 230, 78, 193, 12]

# Create a bar chart with custom colors
plt.bar(x, y, color=['Red', 'Silver', 'Blue', 'White', 'Green'])

# Add labels and title
plt.xlabel('Car Color')
plt.ylabel('Number of Cars Seen')
plt.title('Car Data')

plt.gca().set_facecolor('#f0f0f0')

# Show the plot
plt.show()
```
